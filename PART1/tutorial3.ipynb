```python
#tuple
#ordered
#duplication allowed
#immutable
#hetrogeneous

#-> indexing starts with zero
#-> we can use negative indexing to access the elements from the end of the tuple
#practice question
#t=(10,20,30,40,50)
# 1.we have to print [20,30,40]
t=(10,20,30,40,50)
print(t[1:4])
#2.we have to print [10,20,30]
print(t[:3])
#3.we have to print [30,40,50]
print(t[0:5:2])
#4.we have to print [40,20]
print(t[-2:-5:-2])
#5.we have to print [20,40
print(t[1:4:2])
#6.we have to print [50,40,30,20,10]
print(t[::-1])
```

```python
#set
#unordered
#duplication not allowed
#mutable
#hetrogeneous
#syntax-> set_name={elements}
s={10,20,30,40,50}
#to add an element in the set we can use add function
s.add(60)
#to add multiple elements in the set we can use update function
s.update([70,80])
print(s)
#to remove an element from the set we can use remove function
s.remove(20)
print(s)
#pop function is used to remove a random element from the set randomly
s.pop()
#clear function is used to empty the set
s.clear()
print(s)
```

```python
#dictionary= key-value pair
#unordered
#duplication not allowed
#mutable
#hetrogeneous
#syntax-> dict_name={key:value}
#dictionary is used to compare the data in the form of key-value pair
d={"name":"john","age":30,"city":"new york"}
# to acsess the value of a key we can use the key name
print(d["name"])
#to add a new key-value pair in the dictionary we can use the following syntax
d["country"]="USA"
print(d)
#to remove a key-value pair from the dictionary we can use the pop function
d.pop("age")
print(d)
#to remove a key-value pair from the dictionary we can also use the del function
del d["city"]
print(d)
#to get entire dictionary we can use items function
print(d.items())
#to clear the dictionary we can use the clear function
d.clear()
print(d)
```

```python
#nested dictionary
#dictionary inside a dictionary is called nested dictionary
d={"Guntur":{"pulihara","kothapet"},"Hyderabad":{"kachiguda","charminar"}}
print(d["Guntur"])
```
